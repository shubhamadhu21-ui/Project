```python
# Soil pH Classification Program

ph = float(input("Enter the soil pH value: "))

if ph < 5.5:
    print("Soil Type: Highly acidic")
    print("Recommendation: Needs lime application")

elif 5.5 <= ph < 6.5:
    print("Soil Type: Acidic")
    print("Recommendation: May need amendment (lime)")

elif 6.5 <= ph < 7.5:
    print("Soil Type: Neutral")
    print("Recommendation: Ideal for most crops")

elif 7.5 <= ph < 8.5:
    print("Soil Type: Alkaline")
    print("Recommendation: Needs sulfur application")

else:
    print("Soil Type: Highly alkaline")
    print("Recommendation: Needs major amendment")
```

```python
# Base yield (quintals per hectare)
base_yield = 40

# Inputs
rainfall = float(input("Enter annual rainfall (mm): "))
temperature = float(input("Enter average temperature (°C): "))
soil_quality = float(input("Enter soil quality (0–100): "))

yield_value = base_yield

# ---- Rainfall effect ----
if rainfall > 600:
    increase_steps = (rainfall - 600) // 10
    yield_value += yield_value * (0.02 * increase_steps)

elif rainfall < 400:
    decrease_steps = (400 - rainfall) // 10
    yield_value -= yield_value * (0.02 * decrease_steps)

# ---- Temperature effect ----
if temperature < 25:
    diff = 25 - temperature
    yield_value -= yield_value * (0.01 * diff)

elif temperature > 30:
    diff = temperature - 30
    yield_value -= yield_value * (0.01 * diff)
```
