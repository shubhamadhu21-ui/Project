# 1. Variables and Data Types (9 Examples)

## Concept: Variables store data values. Python uses dynamic typing.

---

```python
# EXAMPLE 1: Simple Variable Assignment
print('='*60)
print('EXAMPLE 1: Simple Variable Assignment')
print('='*60)

name = 'Alice'
print(f"name = '{name}'")
print(f"Type: {type(name)}")
print(f"Type name: {type(name).__name__}")

# Output interpretation: Python identified 'Alice' as a string automatically
```

```python
# EXAMPLE 2: Multiple Data Types
print('='*60)
print('EXAMPLE 2: Multiple Data Types')
print('='*60)

age = 25              # Integer
height = 5.6          # Float
is_student = True     # Boolean
nickname = 'Ali'      # String

print(f'age = {age:20} type = {type(age).__name__}')
print(f'height = {height:20} type = {type(height).__name__}')
print(f'is_student = {str(is_student):20} type = {type(is_student).__name__}')
print(f'nickname = {nickname:20} type = {type(nickname).__name__}')

# Output interpretation: Each variable has different type
```

```python
# EXAMPLE 3: Variable Reassignment (Dynamic Typing)
print('='*60)
print('EXAMPLE 3: Variable Reassignment')
print('='*60)

value = 10
print(f'Step 1: value = {value}, type = {type(value).__name__}')

value = 'Hello'
print(f'Step 2: value = {value}, type = {type(value).__name__}')

value = 3.14
print(f'Step 3: value = {value}, type = {type(value).__name__}')

# Output interpretation: Same variable can hold different types
```

```python
# EXAMPLE 4: Multiple Assignment
print('='*60)
print('EXAMPLE 4: Multiple Assignment')
print('='*60)

x, y, z = 10, 20, 30
print(f'x={x}, y={y}, z={z}')

# Swap using multiple assignment (elegant!)
a, b = 5, 15
print(f'\nBefore swap: a={a}, b={b}')
a, b = b, a
print(f'After swap: a={a}, b={b}')

# Output interpretation: Assigns in order; swap is clean
```

```python
# EXAMPLE 5: Same Value to Multiple Variables
print('='*60)
print('EXAMPLE 5: Same Value Assignment')
print('='*60)

p = q = r = 100
print(f'p={p}, q={q}, r={r}')
print(f'p is q: {p is q}')
print(f'q is r: {q is r}')

# Change one
p = 200
print(f'\nAfter p = 200:')
print(f'p={p}, q={q}, r={r}')

# Output interpretation: Initially same, but independent after change
```

```python
# EXAMPLE 6: Variable Naming Conventions
print('='*60)
print('EXAMPLE 6: Variable Naming Conventions')
print('='*60)

my_name = 'Bob'          # snake_case (recommended)
user_age = 30
is_active = True
_private_var = 'private' # Convention for private
CONSTANT_VALUE = 3.14159 # UPPERCASE for constants

print(f'my_name: {my_name}')
print(f'user_age: {user_age}')
print(f'is_active: {is_active}')
print(f'_private_var: {_private_var}')
print(f'CONSTANT_VALUE: {CONSTANT_VALUE}')

# Output interpretation: Naming conventions make code readable
```

```python
# EXAMPLE 7: Checking Types Programmatically
print('='*60)
print('EXAMPLE 7: Checking All Data Types')
print('='*60)

variables = [
    (42, 'integer'),
    (3.14, 'float'),
    ('hello', 'string'),
    (True, 'boolean'),
    ([1, 2, 3], 'list'),
    ((1, 2), 'tuple'),
    ({'a': 1}, 'dictionary'),
    ({1, 2, 3}, 'set')
]

for var, description in variables:
    print(f'{description:15} : {str(var):20} -> {type(var).__name__}')

# Output interpretation: Python has 8 main built-in types
```

```python
# EXAMPLE 8: Using isinstance() for Type Checking
print('='*60)
print('EXAMPLE 8: Using isinstance() for Type Checking')
print('='*60)

test_value = 42

print(f'Value: {test_value}')
print(f'isinstance(test_value, int): {isinstance(test_value, int)}')
print(f'isinstance(test_value, float): {isinstance(test_value, float)}')
print(f'isinstance(test_value, str): {isinstance(test_value, str)}')

# Check multiple types
print(f'\nisinstance(test_value, (int, float)): {isinstance(test_value, (int, float))}')

# Output interpretation: isinstance() is better than type() for checking
```

```python
# EXAMPLE 9 (COMPLEX): Variable Scope and Memory
print('='*60)
print('EXAMPLE 9 (COMPLEX): Scope and Variable Lifetime')
print('='*60)

global_var = 'I am global'
print(f'Global scope: {global_var}')

def demonstrate_scope():
    local_var = 'I am local'
    print(f'Inside function - can access global_var: {global_var}')
    print(f'Inside function - local_var: {local_var}')
    return local_var

result = demonstrate_scope()
print(f'\nOutside function - global_var: {global_var}')
print(f'Outside function - result: {result}')

print(f'\nTrying to access local_var from outside:')
try:
    print(local_var)
except NameError as e:
    print(f'ERROR: {e}')

# Output interpretation: Local vars exist only in function; global accessible everywhere
```

# 2. Numbers and Arithmetic Operations (9 Examples)

## Concept: Python supports int, float with 7 arithmetic operators: +, -, *, /, //, %, **

---

```python
# EXAMPLE 1: Basic Arithmetic Operations
print('='*60)
print('EXAMPLE 1: Basic Arithmetic Operations')
print('='*60)

a = 20
b = 3

print(f'a = {a}, b = {b}')
print(f'Addition:       {a} + {b} = {a + b}')
print(f'Subtraction:    {a} - {b} = {a - b}')
print(f'Multiplication: {a} * {b} = {a * b}')
print(f'Division:       {a} / {b} = {a / b}')

# Output interpretation: All basic operations work as expected
```

```python
# EXAMPLE 2: Division Types
print('='*60)
print('EXAMPLE 2: Regular vs Floor Division')
print('='*60)

x = 17
y = 5

print(f'x = {x}, y = {y}')
print(f'Regular Division (/) : {x} / {y} = {x / y}')
print(f'Type of / result: {type(x / y).__name__}')

print(f'Floor Division (//): {x} // {y} = {x // y}')
print(f'Type of // result: {type(x // y).__name__}')

print(f'Modulus (%) : {x} % {y} = {x % y}')

# Output interpretation: / gives float, // gives int, % gives remainder
```

```python
# EXAMPLE 3: Exponentiation
print('='*60)
print('EXAMPLE 3: Exponentiation and Roots')
print('='*60)

base = 2
exp = 10
print(f'{base} ** {exp} = {base ** exp}')

num = 5
print(f'\n{num}² = {num ** 2}')
print(f'{num}³ = {num ** 3}')
print(f'{num}⁴ = {num ** 4}')

# Roots using fractional exponents
print(f'\nSquare root of 16: 16**(1/2) = {16 ** (1/2)}')
print(f'Cube root of 27: 27**(1/3) = {27 ** (1/3):.4f}')
print(f'Fourth root of 81: 81**(1/4) = {81 ** (1/4):.4f}')

# Output interpretation: Use ** for powers and roots
```

```python
# EXAMPLE 4: Order of Operations (PEMDAS/BODMAS)
print('='*60)
print('EXAMPLE 4: Order of Operations')
print('='*60)

expr1 = 10 + 5 * 2
expr2 = (10 + 5) * 2
expr3 = 20 - 10 / 2
expr4 = (20 - 10) / 2

print(f'10 + 5 * 2 = {expr1}          (multiply first)')
print(f'(10 + 5) * 2 = {expr2}        (parentheses first)')
print(f'20 - 10 / 2 = {expr3}         (divide first)')
print(f'(20 - 10) / 2 = {expr4}       (parentheses first)')

# Complex expression
complex_expr = 2 + 3 * 4 - 5 / 2 + 1 ** 3
print(f'\n2 + 3 * 4 - 5 / 2 + 1 ** 3 = {complex_expr}')

# Output interpretation: Follow PEMDAS; use () to be explicit
```

```python
# EXAMPLE 5: Augmented Assignment Operators
print('='*60)
print('EXAMPLE 5: Augmented Assignment Operators')
print('='*60)

x = 50
print(f'Initial value: x = {x}')

x += 10
print(f'After x += 10: x = {x}')

x -= 5
print(f'After x -= 5: x = {x}')

x *= 2
print(f'After x *= 2: x = {x}')

x //= 3
print(f'After x //= 3: x = {x}')

x %= 10
print(f'After x %= 10: x = {x}')

# Output interpretation: Shorthand for x = x operation
```

```python
# EXAMPLE 6: Working with Floats
print('='*60)
print('EXAMPLE 6: Floating-Point Numbers')
print('='*60)

a = 10.5
b = 3.2

print(f'a = {a}, b = {b}')
print(f'a + b = {a + b}')
print(f'a - b = {a - b}')
print(f'a * b = {a * b}')
print(f'a / b = {a / b}')

# Precision issue
print(f'\nFloating-point precision:')
print(f'0.1 + 0.2 = {0.1 + 0.2}')
print(f'0.1 + 0.2 == 0.3: {0.1 + 0.2 == 0.3}')
print(f'Is close? {abs((0.1 + 0.2) - 0.3) < 1e-9}')

# Output interpretation: Floats have precision limits
```

```python
# EXAMPLE 7: Built-in Math Functions
print('='*60)
print('EXAMPLE 7: Built-in Math Functions')
print('='*60)

print(f'abs(-15) = {abs(-15)}')
print(f'abs(15) = {abs(15)}')
print(f'abs(-3.14) = {abs(-3.14)}')

print(f'\nround(3.14159, 2) = {round(3.14159, 2)}')
print(f'round(9.7) = {round(9.7)}')
print(f'round(9.5) = {round(9.5)}')

print(f'\nmin(10, 5, 20, 3) = {min(10, 5, 20, 3)}')
print(f'max(10, 5, 20, 3) = {max(10, 5, 20, 3)}')

print(f'\npow(2, 3) = {pow(2, 3)}')
print(f'pow(2, 0.5) = {pow(2, 0.5):.4f}')

# Output interpretation: Python has many useful built-in functions
```

```python
# EXAMPLE 8: Practical Use - Modulo and Division
print('='*60)
print('EXAMPLE 8: Practical Applications')
print('='*60)

# Convert seconds to hours:minutes:seconds
total_seconds = 3665
hours = total_seconds // 3600
remaining = total_seconds % 3600
minutes = remaining // 60
seconds = remaining % 60

print(f'Total seconds: {total_seconds}')
print(f'Converted to: {hours}h {minutes}m {seconds}s')

# Check if number is even or odd
print(f'\nEven/Odd Check:')
for num in [7, 12, 25, 100]:
    if num % 2 == 0:
        print(f'  {num} is EVEN')
    else:
        print(f'  {num} is ODD')

# Output interpretation: Modulo is useful for cycling, remainders
```

```python
# EXAMPLE 9 (COMPLEX): Real-world Calculations
print('='*60)
print('EXAMPLE 9 (COMPLEX): Compound Interest Calculator')
print('='*60)

# Compound Interest Formula: A = P(1 + r/n)^(nt)
principal = 1000  # Initial investment in dollars
annual_rate = 0.05  # 5% per year
time_years = 3
compounding_periods = 12  # Monthly

amount = principal * (1 + annual_rate/compounding_periods) ** (compounding_periods * time_years)
interest = amount - principal

print(f'Principal: ${principal:.2f}')
print(f'Annual Rate: {annual_rate * 100}%')
print(f'Time: {time_years} years')
print(f'Compounding: Monthly')
print(f'\nFinal Amount: ${amount:.2f}')
print(f'Interest Earned: ${interest:.2f}')

# BMI Calculation: BMI = weight(kg) / height(m)^2
print(f'\n{'='*60}')
print(f'BMI Calculator')
print(f'{'='*60}')

weight_kg = 70
height_m = 1.75

bmi = weight_kg / (height_m ** 2)
print(f'Weight: {weight_kg} kg, Height: {height_m} m')
print(f'BMI = {bmi:.2f}')

if bmi < 18.5:
    category = 'Underweight'
elif bmi < 25:
    category = 'Normal weight'
elif bmi < 30:
    category = 'Overweight'
else:
    category = 'Obese'

print(f'Category: {category}')

# Output interpretation: Real applications of arithmetic operations
```
